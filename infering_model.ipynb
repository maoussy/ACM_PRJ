```python
import torch
from torchvision import models

# Define the base model structure
model = models.resnet152(pretrained=False)  # Create a ResNet-152 model without pre-trained weights

# Modify the output size of the final layer to match the number of classes in your dataset
num_classes = 18
model.fc = torch.nn.Linear(model.fc.in_features, num_classes)

# Load the state dictionary
state_dict = torch.load(r"C:\Users\Jean-Michel\Desktop\Models\tsr_resnet152_25_9_10PM.pt")

# Load the state dictionary into the model structure
model.load_state_dict(state_dict)
model.eval()  # Set the model to evaluation mode
```

```python
from torchvision import transforms
from torch.utils.data import DataLoader
from torchvision.datasets import ImageFolder
import torch

# Ensure you have a CUDA device available
if not torch.cuda.is_available():
    raise RuntimeError("CUDA is not available")
```

```python
# Transfer your model to the GPU
model = model.to('cuda')

transform = transforms.Compose([
    transforms.Resize((224, 224)),  
    transforms.ToTensor(),  
])

test_dataset = ImageFolder(root=r"C:\Users\Jean-Michel\Desktop\data_updated", transform=transform)
data_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

correct = 0
total = 0 

with torch.no_grad():  
    for inputs, labels in data_loader:
       
        inputs = inputs.to('cuda')
        labels = labels.to('cuda')
        
        
        outputs = model(inputs)
 # Get the predicted class for each sample
        _, predicted = torch.max(outputs.data, 1)
        
        # Update the total number of samples and the number of correct predictions
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

# Print the accuracy
print(f'Accuracy: {100 * correct / total}%')      
```
